```python
import os
import json
import pandas as pd
import sys

subjects=[f"A0{i+1}" for i in range(9)]
home_dir = os.getcwd()
project_root = os.path.dirname(os.path.dirname(home_dir))
sys.path.append(project_root)

from src.utils.config import ACQ_FUNC_LIST, FREQ_BANDS_LIST

print(ACQ_FUNC_LIST, FREQ_BANDS_LIST)
```

```python
result = []
for subject in subjects:
    for min_freq,max_freq in FREQ_BANDS_LIST:
        for acq_func in ACQ_FUNC_LIST:
            result_dir = os.path.abspath(os.path.join(home_dir, f"../bayescspsvm/{subject}/TPE/{acq_func}_{min_freq}_{max_freq}"))
            file= os.path.join(result_dir,"best_eeg_params.json")
            with open(file, 'r') as file:
                data = json.load(file)
            data['subject']=subject
            data['min_freq']=min_freq
            data['max_freq']=max_freq
            result.append(data)
df = pd.DataFrame(result)
df
```

```python
result1=df[[ 'subject', 'acq_func', 'min_freq', 'max_freq', 't_start', 't_end', 'n_components', 'C', 'kernel', 'gamma', 'accuracy', 'kappa']]
result1=result1.rename(columns={'accuracy': 'train_acc'})
result1
```

```python
# Result 3
result3 = pd.DataFrame()
for subject in subjects:
    temp = result1[result1['subject']==subject]
    result3=pd.concat([result3, result1.iloc[[temp['kappa'].idxmax()],:]])
result3
```
